# Import DataSet


```python
from roboflow import Roboflow

# Initialize Roboflow
rf = Roboflow(api_key="RoboFlow Api")  # replace with your API key
```

```python
# Replace with your workspace and project names
workspace_name = "smarthsrp"
project_name = "license-plate-recognition-rxg4e-xyod6"
version_number = 2  # the version you created/forked

project = rf.workspace(workspace_name).project(project_name)
dataset = project.version(version_number).download("yolov8")  # download in YOLOv8 format use same format in yolov10

print("Dataset downloaded to:", dataset.location)
```

```python
# 4ï¸âƒ£ Set paths
dataset_path = "/content/License-Plate-Recognition-2"
data_yaml = os.path.join(dataset_path, "data.yaml")
project_folder = "/content/plate_recognition_project"
run_name = "yolov10_plate"
```

```python
print("Train images:", len(os.listdir(os.path.join(dataset_path, "train/images"))))
print("Validation images:", len(os.listdir(os.path.join(dataset_path, "valid/images"))))
print("Test images:", len(os.listdir(os.path.join(dataset_path, "test/images"))))
```

# Model checkpoint is Saved in My Drive

```python
from google.colab import drive   # import os is also required
# 1ï¸âƒ£ Mount Google Drive
drive.mount('/content/drive')

# 2ï¸âƒ£ Paths
drive_path = "/content/drive/MyDrive/Licence_Plate"
os.makedirs(drive_path, exist_ok=True)
```

# Load Model

```python
model=YOLO("/content/drive/MyDrive/Licence_Plate/Licence_Plate_1/weights/last.pt") # Path of the the last.pt or the  last epoch checkpoint i.e epoch50
```

# Train model

```python
model.train(
    data=data_yaml,
    epochs=100,
    imgsz=640,
    batch=16,
    #augment=True,
    #mosaic=1.0,            # add mosaic now
    #mixup=0.5,             # add mixup now
    #perspective=0.5,
    project=drive_path,
    name="Licence_Plate_1",
    exist_ok=True,
    save_period=5,
    resume=True
)
```

# Result of Validation Set
P (Precision) = 0.981 â†’ Very few false positives

R (Recall) = 0.949 â†’ Most objects detected

mAP50 = 0.979 â†’ Mean Average Precision at 50% IoU, excellent

mAP50-95 = 0.72 â†’ Average over stricter IoU thresholds, still decent

# Run test evaluation on test dataset

```python
model = YOLO("/content/drive/MyDrive/Licence_Plate/Licence_Plate_1/weights/best.pt")

metrics = model.val(
    data=data_yaml,
    split="test",
    imgsz=640
)

print(metrics)
```
